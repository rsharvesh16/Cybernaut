**Task 2 - Plotting data using Seaborn on Kaggle Crop Production Data Set**


```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
```

```python
df=pd.read_csv("crop_production[1].csv")
df[:5]
```

```python
plt.hist(df['Season'])
```

```python
sns.lineplot(df["Crop_Year"],df["Production"])
```

```python
sns.barplot(df["State_Name"],df["Area"])
plt.xticks(rotation=90)
```

```python
sns.countplot(x=df['Crop_Year'])
plt.xticks(rotation=90)
```
